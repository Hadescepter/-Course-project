```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#定义初始点
x=1
y=1
l=0.1
```

```python
def func_xy(x,y):
    return x*x+2*y*y-2*x*y
```

```python
def gradient_x(x,y):
    return 2*x-2*y
def gradient_y(x,y):
    return 4*y-2*x
```

```python
result_v=[]
result_x=[]
result_y=[]
result_T=[]
for l in [0.1,0.01,0.001]:
    value=[]
    x_list=[]
    y_list=[]
    time_list=[]
    x=1
    y=1
    for epoch in range(1000):
        value.append(func_xy(x,y))
        fx=gradient_x(x,y)
        fy=gradient_y(x,y)
        x_list.append(x)
        y_list.append(y)
        time_list.append(epoch)
        if fx == 0 and fy == 0:
          break
        x-=l*fx
        y-=l*fy
    result_v.append(value)
    result_x.append(x_list)
    result_y.append(y_list)
    result_T.append(time_list)
```

```python
plt.plot(np.array(result_T[0]),np.array(result_v[0]),label='learning rate=0.1')
plt.plot(np.array(result_T[1]),np.array(result_v[1]),label='learning rate=0.01')
plt.plot(np.array(result_T[2]),np.array(result_v[2]),label='learning rate=0.001')
#显示图例
plt.ylabel('function_value')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
plt.plot(np.array(result_T[0]),np.array(result_x[0]),label='learning rate=0.1')
plt.plot(np.array(result_T[1]),np.array(result_x[1]),label='learning rate=0.01')
plt.plot(np.array(result_T[2]),np.array(result_x[2]),label='learning rate=0.001')
#显示图例
plt.ylabel('x_value')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
plt.plot(np.array(result_T[0]),np.array(result_y[0]),label='learning rate=0.1')
plt.plot(np.array(result_T[1]),np.array(result_y[1]),label='learning rate=0.01')
plt.plot(np.array(result_T[2]),np.array(result_y[2]),label='learning rate=0.001')
#显示图例
plt.ylabel('y_value')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
