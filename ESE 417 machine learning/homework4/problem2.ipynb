```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from collections import Counter
```

```python
digits = load_digits()
X, Y = digits.data, digits.target
print(X.shape,Y.shape)
```

```python
counter = Counter(Y)
labels = list(counter.keys())
num = list(counter.values())
plt.figure(figsize=(8, 6))
plt.bar(labels, num, color='green',tick_label=labels,)
plt.xlabel('Labels')
plt.ylabel('Num of Labels')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.3,random_state=55)
```

```python
hidden_nodes_sizes=np.arange(10,110,5)
various_hidden_nodes_scores={}
for hidden_node_size in hidden_nodes_sizes:
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(hidden_node_size, 10), random_state=55)
    clf.fit(x_train,y_train)
    y_predict=clf.predict(x_test)
    accuracy=accuracy_score(y_test,y_predict)
    various_hidden_nodes_scores[hidden_node_size]=accuracy

#输出accuracy
print("Test Accuracies for Different hidden_nodes_size:")
for size, accuracy in various_hidden_nodes_scores.items():
    print(f"{size}: {accuracy:.2f}")

    
```

```python
#画出准确率结果

plt.figure(figsize=(8,6))
plt.plot(various_hidden_nodes_scores.keys(),various_hidden_nodes_scores.values(),color='green')
plt.title("The Scores of Different Hidden_nodes_size") 
plt.xlabel("hidden_nodes_sizes(n,10)")
plt.ylabel("accuracy")
```
